```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Load the dataset
data = pd.read_csv('Boston.csv')
```

```python
# Explore the dataset
print(data.head())
```

```python
# Visualize the distribution of the target variable
plt.figure(figsize=(8, 6))
sns.histplot(data['medv'], kde=True)
plt.title('Distribution of the Target Variable')
plt.xlabel('MEDV')
plt.show()
```

```python
# Split the dataset into training and test sets
x = data[['rm']]
y = data['medv']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
# Fit the linear regression model on the training data
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
# Predict the target variable for the training and test data
y_pred_train = regressor.predict(x_train)
y_pred_test = regressor.predict(x_test)
```

```python
# Evaluate the performance of the model on the training and test data
print('Coefficients:', regressor.coef_)
print('Intercept:', regressor.intercept_)
print('Training R2 Score:', regressor.score(x_train, y_train))
print('Test R2 Score:', regressor.score(x_test, y_test))
```

```python
# Visualize the regression line
plt.figure(figsize=(8, 6))
sns.scatterplot(x=x_train['rm'], y=y_train, color='blue')
plt.plot(x_train['rm'], y_pred_train, color='red', linewidth=2)
plt.xlabel('Average Number of Rooms per Dwelling')
plt.ylabel('Median Value of Owner-Occupied Homes')
plt.title('Linear Regression Fit on Training Data')
plt.show()
```
